```python
from scipy.io import loadmat
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

data = loadmat('ex8data1.mat')
X, y = data['Xval'], data['yval']
```

```python
def get_mv(X):
    mu, sigma = np.empty(X.shape[1]), np.empty(X.shape[1])
    mu = np.mean(X, axis=0)
    var = np.var(X, axis=0)
    return mu, var

mu, var = get_mv(X)
```

```python
xplot = np.linspace(0, 25, 100)
yplot = np.linspace(0, 25, 100)
Xplot, Yplot = np.meshgrid(xplot, yplot)
Z = 1/(2*np.pi*(var[0]*var[1])**0.5)*np.exp(-((Xplot-mu[0])**2)/(2*var[0])-((Yplot-mu[1])**2)/(2*var[1]))
plt.figure(figsize=(12,8))
plt.scatter(X[:,0], X[:, 1])
plt.contour(Xplot, Yplot, Z, [10**-11, 10**-7, 10**-5, 10**-3, 0.01], color='k')
```

```python
dist1 = stats.norm(mu[0], var[0]).pdf(X[:,0])
dist2 = stats.norm(mu[1], var[1]).pdf(X[:,1])
threshold = np.linspace(0, 1, 1000)
dist = dist1*dist2

def get_epsilon(X, y):
    best_f1 = 0
    b_epsilon = 0
    for epsilon in np.linspace(0, 1, 1000):
        pred = X < epsilon
        tp = np.sum(np.logical_and(pred==1, y==1)).astype(float)
        fn = np.sum(np.logical_and(pred==0, y==1)).astype(float)
        fp = np.sum(np.logical_and(pred==1, y==0)).astype(float)
        tn = np.sum(np.logical_and(pred==0, y==0)).astype(float)
        if tp == 0:
            f1 = 0
        else:
            precision = tp/(tp+fp)
            recall = tp/(tp+fn)
            f1 = 2*precision*recall/(precision+recall)
        if f1 > best_f1:
            best_f1 = f1
            b_epsilon = epsilon
    return b_epsilon, best_f1
epsilon, f1 = get_epsilon(dist, y.ravel())

neg = X[dist>epsilon]
pos = X[dist<epsilon]
plt.figure(figsize=(12,8))
plt.scatter(neg[:,0], neg[:,1])
plt.scatter(pos[:,0], pos[:,1])
f1, epsilon
```

```python
from scipy.io import loadmat
from scipy.optimize import minimize
import numpy as np

data = loadmat('ex8_movies.mat')
data2 = loadmat('ex8_movieParams.mat')
Y, R = data['Y'], data['R']
nu = data2['num_users'][0,0]
nm = data2['num_movies'][0,0]
nf = data2['num_features'][0,0]
para = np.random.random((nm+nu)*nf)
lam = 1
Y = Y-Y.mean()
# X, theta = data2['X'], data2['Theta']
# param = np.concatenate((X.ravel(), theta.ravel()))

def cost(para, y):
    para = np.matrix(para)
    y = np.matrix(y)
    X = para[0, :nm*nf].reshape((nm, nf))
    theta = para[0, nm*nf:].reshape((nu, nf))
    
    return 0.5*np.power(np.multiply(X@theta.T-y, R),2).sum()

def cost_reg(para, y, lam):
    para = np.matrix(para)
    y = np.matrix(y)
    X = para[0, :nm*nf].reshape((nm, nf))
    theta = para[0, nm*nf:].reshape((nu, nf))
    Cost = cost(para, y)
    reg = lam/2*(np.sum(np.power(theta, 2))+np.sum(np.power(X, 2)))
    return Cost+reg

def gradient(para, y, lam):
    para = np.matrix(para)
    y = np.matrix(y)
    X = para[0, :nm*nf].reshape((nm, nf))
    theta = para[0, nm*nf:].reshape((nu, nf))
    
    A = np.multiply(X@theta.T-y, R)
    B = A@theta
    C = A.T@X
    grad = np.concatenate((B.ravel(), C.ravel()), axis=1)
    reg = lam*np.concatenate((X.ravel(), theta.ravel()), axis=1)
    return grad+reg

result = minimize(fun=cost_reg, x0=para, method='TNC', args=(Y, lam), jac=gradient)
```

```python
para = result['x']
X = para[:nm*nf].reshape((nm, nf))
theta = para[nm*nf:].reshape((nu, nf))
```

```python
pred = X@theta.T
recommend_list = []
for user in range(pred.shape[1]):
    recommend_list.append(pred[:, user][R[:, user]==0].argmax())
```
