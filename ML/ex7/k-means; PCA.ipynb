```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from scipy.io import loadmat

img = mpimg.imread('bird_small.png')
data = loadmat('bird_small.mat')['A']
data_re = data.reshape((data.shape[0]*data.shape[1], data.shape[2]))/255

iters = 10
eps = 1e-2
k = 60
plt.imshow(img)
plt.show()

def k_means(k, X):
    idx = np.random.choice(X.shape[0], k, replace=False)
    centroid = [X[i:i+1,:] for i in idx]
    centroid_s = np.repeat(centroid, X.shape[0], axis=1)
    XX = np.repeat([X], k, axis=0)
    while True:
        distance = np.sum(np.power(XX-centroid_s, 2), axis=2)
        nearest_centroid = np.argmin(distance, axis=0)
        centroid_update = []
        for i in range(k):
            idx = [j for j in range(X.shape[0]) if nearest_centroid[j] == i]
            centroid_update.append([np.sum(X[idx, :], axis=0)/len(idx)])
        centroid_update = np.array(centroid_update)
        define = np.sum(np.power(centroid-centroid_update, 2), axis=2).reshape(k)
        if (define > eps).any():
            centroid = centroid_update
            continue
        else:
            break
    return nearest_centroid, centroid

def run_k_means(iters):
    error_min = 10000
    nearest_f = []
    centroid_f = np.empty((k,1, 3))
    for i in range(iters):
        error = 0
        nearest, centroid = k_means(k, data_re)
        for i in range(k):
            idx = [j for j in range(data_re.shape[0]) if nearest[j] == i]
            error += np.sum(np.power(data_re[idx, :]-centroid[i], 2))/len(idx)
        if error < error_min:
            error_min = error
            nearest_f = nearest
            centroid_f = centroid
    return nearest_f, centroid_f

nearest, centroid = run_k_means(iters)
centroid = np.array(centroid)
X_recover = centroid[nearest, 0, :].reshape((data.shape[0], data.shape[1], data.shape[2]))
plt.imshow(X_recover)
```

```python
from sklearn.cluster import KMeans

data = loadmat('bird_small.mat')['A']
data_re = data.reshape((data.shape[0]*data.shape[1], data.shape[2]))/255

model = KMeans(n_clusters=16)
model.fit(data_re)
centroid = model.cluster_centers_
idx = model.predict(data_re)

X_recover = centroid[idx, :].reshape((data.shape[0], data.shape[1], data.shape[2]))
plt.imshow(X_recover)
```

```python
import numpy as np
from scipy.io import loadmat
import matplotlib.pyplot as plt

data = loadmat('ex7faces.mat')
X = data['X']

X_chosen = X[3,:].reshape((32,32)).T
plt.matshow(X_chosen)
plt.xticks(np.array([]))
plt.yticks(np.array([]))
plt.show()
mean = []
std = []
for i in range(X.shape[1]):
    m = X[:,i].mean()
    s = X[:,i].std()
    X[:,i] = (X[:,i]-m)/s
    mean.append(m)
    std.append(s)

cov = X.T@X/X.shape[0]
U, S, V = np.linalg.svd(cov)
X_pca = X@U[:, 0:100]
X_recover = X_pca@U[:, 0:100].T
for i in range(X_recover.shape[1]):
    X_recover[:,i] = X_recover[:,i]*std[i]+mean[i]

plt.matshow(X_recover[3,:].reshape((32,32)).T)
plt.xticks(np.array([]))
plt.yticks(np.array([]))
plt.show()
```
